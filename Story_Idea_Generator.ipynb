```python
import google.generativeai as genai
import gradio as gr
```

```python
genai.configure(api_key="AIzaSyAi7NCoWVdV40h9Ag0I4ASKwMVvOGSDZUo")
```

```python
def generate_story_prompt(genre, theme, character):
    user_prompt = f"""Generate a creative story idea prompt.
       Genre : {genre}
       Theme : {theme}
       Main Character : {character}
       """
    model = genai.GenerativeModel("gemini-1.5-flash")
    response = model.generate_content(user_prompt)
    return response.text
```

```python
iface = gr.Interface(
    fn=generate_story_prompt,
    inputs=[
        gr.Textbox(label="Enter Genre"),
        gr.Textbox(label="Enter Theme"),
        gr.Textbox(label="Enter Main Character"),
    ],
    outputs=gr.Textbox(label="Generated Story Idea"),
    title="ðŸ“– Story Idea Generator",
    description="Enter a genre, theme, and main character to get a creative story idea."
)

# Launch app
iface.launch()
```
